```python
import tkinter as tk
from tkinter import messagebox

class ATM:
    def __init__(self, user_id, pin):
        self.user_id = user_id
        self.pin = pin
        self.balance = 0
        self.transaction_history = []

    def show_menu(self):
        menu = tk.Toplevel()
        menu.title("ATM Menu")
        menu.geometry("300x200")

        label = tk.Label(menu, text="Choose an option:")
        label.pack()

        view_history_btn = tk.Button(menu, text="View Transaction History", command=self.view_transaction_history)
        view_history_btn.pack()

        withdraw_btn = tk.Button(menu, text="Withdraw Money", command=self.withdraw_prompt)
        withdraw_btn.pack()

        deposit_btn = tk.Button(menu, text="Deposit Money", command=self.deposit_prompt)
        deposit_btn.pack()

        transfer_btn = tk.Button(menu, text="Transfer Money", command=self.transfer_prompt)
        transfer_btn.pack()

        quit_btn = tk.Button(menu, text="Quit", command=menu.destroy)
        quit_btn.pack()

    def view_transaction_history(self):
        history = tk.Toplevel()
        history.title("Transaction History")
        history.geometry("300x200")

        label = tk.Label(history, text="Transaction History:")
        label.pack()

        for transaction in self.transaction_history:
            transaction_label = tk.Label(history, text=transaction)
            transaction_label.pack()

    def withdraw_prompt(self):
        withdraw_window = tk.Toplevel()
        withdraw_window.title("Withdraw Money")
        withdraw_window.geometry("300x200")

        label = tk.Label(withdraw_window, text="Enter amount to withdraw:")
        label.pack()

        amount_entry = tk.Entry(withdraw_window)
        amount_entry.pack()

        confirm_btn = tk.Button(withdraw_window, text="Withdraw", command=lambda: self.withdraw(float(amount_entry.get())))
        confirm_btn.pack()

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            self.transaction_history.append(f"Withdrawal of ${amount}")
            messagebox.showinfo("Success", "Withdrawal successful.")
        else:
            messagebox.showerror("Error", "Insufficient balance.")

    def deposit_prompt(self):
        deposit_window = tk.Toplevel()
        deposit_window.title("Deposit Money")
        deposit_window.geometry("300x200")

        label = tk.Label(deposit_window, text="Enter amount to deposit:")
        label.pack()

        amount_entry = tk.Entry(deposit_window)
        amount_entry.pack()

        confirm_btn = tk.Button(deposit_window, text="Deposit", command=lambda: self.deposit(float(amount_entry.get())))
        confirm_btn.pack()

    def deposit(self, amount):
        self.balance += amount
        self.transaction_history.append(f"Deposit of ${amount}")
        messagebox.showinfo("Success", "Deposit successful.")

    def transfer_prompt(self):
        transfer_window = tk.Toplevel()
        transfer_window.title("Transfer Money")
        transfer_window.geometry("300x200")

        recipient_label = tk.Label(transfer_window, text="Enter recipient's User ID:")
        recipient_label.pack()

        recipient_entry = tk.Entry(transfer_window)
        recipient_entry.pack()

        amount_label = tk.Label(transfer_window, text="Enter amount to transfer:")
        amount_label.pack()

        amount_entry = tk.Entry(transfer_window)
        amount_entry.pack()

        confirm_btn = tk.Button(transfer_window, text="Transfer", command=lambda: self.transfer(float(amount_entry.get()), recipient_entry.get()))
        confirm_btn.pack()

    def transfer(self, amount, recipient_id):
        recipient = None
        for u in users:
            if u.user_id == recipient_id:
                recipient = u
                break
        if recipient:
            if self.balance >= amount:
                self.balance -= amount
                recipient.balance += amount
                self.transaction_history.append(f"Transfer of ${amount} to {recipient_id}")
                messagebox.showinfo("Success", "Transfer successful.")
            else:
                messagebox.showerror("Error", "Insufficient balance.")
        else:
            messagebox.showerror("Error", "Recipient not found.")

def create_account():
    user_id = user_id_entry.get()
    pin = pin_entry.get()
    user = ATM(user_id, pin)
    users.append(user)
    messagebox.showinfo("Success", "Account created successfully.")

def login():
    user_id = user_id_entry.get()
    pin = pin_entry.get()
    for user in users:
        if user.user_id == user_id and user.pin == pin:
            user.show_menu()
            return
    messagebox.showerror("Error", "Invalid User ID or PIN. Please try again.")

users = []

root = tk.Tk()
root.title("ATM System")
root.geometry("300x200")

user_id_label = tk.Label(root, text="User ID:")
user_id_label.pack()

user_id_entry = tk.Entry(root)
user_id_entry.pack()

pin_label = tk.Label(root, text="PIN:")
pin_label.pack()

pin_entry = tk.Entry(root, show="*")
pin_entry.pack()

create_account_btn = tk.Button(root, text="Create Account", command=create_account)
create_account_btn.pack()

login_btn = tk.Button(root, text="Login", command=login)
login_btn.pack()

root.mainloop()
```
